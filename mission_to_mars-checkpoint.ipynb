```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
url = 'https://redplanetscience.com/'  
driver = webdriver.Chrome('C:/Users/nella/BOOTCAMP/HOMEWORK/web-scraping-challenge/chromedriver') 
driver.get(url) 

# Retrieve snapshot of website
time.sleep(5) 
html = driver.page_source
```

```python
# Translate it
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
latest_title = soup.find('div', class_='content_title').text
latest_date = soup.find('div', class_='list_date').text
latest_blurb = soup.find('div', class_='article_teaser_body').text

print(latest_date)
print(latest_title)
print(latest_blurb)
```
